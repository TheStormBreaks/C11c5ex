```python
# ARITHEMETIC OPERATORS
a = 10
b = 20
add = (a+b)
diff = (a-b)
product = (a*b)
division = (a/b)
mod = (a%b)
fd = (a//b)
exp = (a**b)

print("Sum", add) # addition 

print("Difference", diff) # subtraction

print("Product", product) # multiplication

print("DIvision", division) 
# divides the operand on the left 
# by operant on the right and
# returns the quotient 

print("Modulus", mod)
# divides the operand on the left 
# by the operand on the right and
# returns the remainder

print("Floor Division", fd)
# divides the operand on the left
# by the operand on the right and
# returns the quotient removing the decimal part
# also called INTEGER DIVISION

print("Exponent", exp)
# raise the operand on the left
# to the power of operand on the right
```

```python
# RELATIONAL OPERATORS
num1 = 10
num2 = 0
num3 = 10
str1 = "good"
str2 = "life"

# Equals to
print(num1 == num2)
print(str1 == str2)

# Not equal to
print(num1 != num2)
print(str1 != str2)
print(num1 != num3)

# Greater Than
print(num1 > num2)
print(str1 > str2)

# Less Than
print(num1 < num3)
print(str2 < str1)

# Greater than or equal to
print(num1 >= num2)
print(num2 >= num3)
print(str1 >= str2)

#Less then or equal to
print(num1 <= num2)
print(num2 <= num3)
print(str1 <= str2)
```

```python
# ASSIGNMENT OPERATORS 


# Assigns value from right
# operator to left operator
num1 = 2
num2 = num1
print(num2)


# Adds the value of right hand side
# operand and asigns the result 
# to left hand side operand
num1 = 10
num2 = 2
num1 += num2
print(num1)
# POINT: x += y is same as x = x + y


# Subtracts the value of right operand
# from the left operand and assigns 
# the result to left operand
num1 = 10
num2 = 2
num1 -= num2
print(num1)
# POINT: x -=y is same as x = x - y


# Multiplies the value of left operand
# by the value of right operand and
# assigns the result to left operand
num1 = 2
num2 = 3
num1 *= num2
print(num1)

# Divides the value of left operand
# by the value of right operand and
# assigns the result to left operand
num1 = 6
num2 = 3
num1 /= num2
print(num1)


# Performs modulus operation usind two operands
# and assigns the result to left operand
num1 = 7
num2 = 3
num1 %= num2
print(num1)
# POINT: x %= y is same as x = x % y


# It performs floor division using two operands
# and assigns the result to the left operand
num1 = 7
num2 = 3
num1 //= num2
print(num1)
# POINT: x //= y is same as x = x // y


# It performs exponential power calculation 
# on operators and assigns value to left operator
num1 = 2
num2 = 3
num1 **= num2
print(num1)
# POINT: x **= y is same as x = x ** y
```
